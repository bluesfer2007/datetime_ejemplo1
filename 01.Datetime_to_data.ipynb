## FORMATEO DE FECHA EN UN DF 

```python
df=pd.read_excel('bdd_facturas.xlsx', skiprows=7)
```

```python
df.info()
```

```python
df.head()
```

```python
#filtrar valores nulos para evitar el error de formateo de fecha
df=df[df['EMISOR'].notnull()]
```

```python
df['FECHA']=df['FECHA'].apply(lambda x: datetime.fromisoformat(str(x)))
```

```python
df.info()
```

```python
df['DIF FECHAS']=df['F. VENCIMIENTO']-df['F. EMISION']
```

```python
df['DIF FECHAS']=df['DIF FECHAS'].apply(lambda x: x.days)
```

```python
df.head()
```

## DIFERENCIA ENTRE LA FECHA ACTUAL Y LA FECHA EMISION

```python
date.today()
```

```python
df['DIAS_TRASNCURRIDOS']=df['F. EMISION'].apply(lambda x: (date.today() -datetime.date(x)).days)
```

```python
df.head()
```

### FORMATEAR A TEXTO LAS FECHAS

```python
locale.setlocale(locale.LC_TIME, '')
```

```python
df['NOMBRE_MES']=df['FECHA'].apply(lambda x: datetime.date(x).strftime('%B %d, %Y'))
```

```python
df.head()
```

## SEGUNDO EJEMPLO CON VALORES DE TIPO HORA

```python
dh=pd.read_csv('Sales_April_2019.csv')
```

```python
dh.head()
```

```python
dh=dh[dh['Order Date'].notnull()]
dh=dh[dh['Order Date']!="Order Date"]
```

```python
#dh['Order Date']=dh['Order Date'].apply(lambda x: datetime.fromisoformat(x))

dh['Order Date']=dh['Order Date'].apply(lambda x: pd.to_datetime(x))
```

```python
dh['YEAR']=dh['Order Date'].apply(lambda x: x.year)
dh['MONTH']=dh['Order Date'].apply(lambda x: x.month)
dh['DAY']=dh['Order Date'].apply(lambda x: x.day)
dh['HOUR']=dh['Order Date'].apply(lambda x: x.hour)
dh['MINUTE']=dh['Order Date'].apply(lambda x: x.minute)
```

```python
dh.head()
```

```python
#fechas personalizadas para comparar
d=date(2019,4,19)
t=time(12,30)
fecha_contraste=datetime.combine(d,t)
fecha_contraste
```

```python
dh['VALIDAR HORA']=dh['Order Date'].apply(lambda x: x>=fecha_contraste)
```
